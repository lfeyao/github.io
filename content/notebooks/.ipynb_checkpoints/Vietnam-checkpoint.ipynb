```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import wbdata
```

```python
countries = ['VNM', 'USA', 'CHN', 'JPN', 'KOR', 'BRA', 'IND']
```

```python
indicators = {'SP.POP.DPND.OL':'Age Dependency Ratio', 'SP.DYN.CBRT.IN': 'Birth rate (per 1,000 people)'}
```

```python
df = wbdata.get_dataframe(indicators, country=countries, convert_date=False)
```

```python
dfu = df.unstack(level=0)
sdfu = dfu[40:]
# sdfu.head()
```

```python
sdfu['Age Dependency Ratio'].plot()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) 
plt.title('Age dependency ratio (Old/Working)')
plt.xlabel('Date')
plt.ylabel('Ratio of old (>64) to the working-age population')
```

```python
sdfu['Birth rate (per 1,000 people)'].plot()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) 
plt.title('Birth Rate')
plt.xlabel('Date')
plt.ylabel('Birth rate (per 1,000 people)')
```
